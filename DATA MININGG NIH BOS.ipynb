### Melakukan pre-processsing data

1. Mengecek apakah ada sel yang kosong pada dataset yang telah diimport.
2. Melakukan pembersihan data dengan mengisi missing value (jika hasil print menampilkan bahwa data perlu diisi missing valuenya).
3. Mengubah tipe data untuk optimalisasi memori.
4. Normalisasi data
5. Menghapus atau menggabungkan kolom yang tidak penting.

```
#cek apakah ada sel data yang kosong
print(df_KorbanJakarta.isnull().sum())
print(df_KorbanPerempuanJenisKejahatan2022.isnull().sum())
print(df_KorbanPerempuanJenisKejahatan2023.isnull().sum())
print(df_KorbanAnakJenisKejahatan2023.isnull().sum())
print(df_KorbanAnakLakilaki2017.isnull().sum())
print(df_KorbanAnakPerempuan2017.isnull().sum())
print(df_KekerasanTempatKejadianJawaTengah.isnull().sum())
print(df_KorbanJenisKejahatanJawaBarat.isnull().sum())
print(df_KekerasanTempatKejadianJakarta.isnull().sum())
print(df_KekerasanTempatKejadianJawaBarat.isnull().sum())
```

```
#mengisi missing value
df_KorbanPerempuanJenisKejahatan2022[['Fisik', 'Psikis', 'Seksual', 'Eksploitasi', 'TPPO', 'Penelantaran', 'Lainnya']] = df_KorbanPerempuanJenisKejahatan2022[['Fisik', 'Psikis', 'Seksual', 'Eksploitasi', 'TPPO', 'Penelantaran', 'Lainnya']].fillna(0)
df_KorbanPerempuanJenisKejahatan2022['Satuan'] = df_KorbanPerempuanJenisKejahatan2022['Satuan'].fillna('Orang')
df_KorbanPerempuanJenisKejahatan2023['Satuan'] = df_KorbanPerempuanJenisKejahatan2023['Satuan'].fillna('Orang')
df_KekerasanTempatKejadianJakarta['kategori_lokasi'] = df_KekerasanTempatKejadianJakarta['kategori_lokasi'].fillna(0)
```

```
#mengubah tipe data untuk optimalisasi memori
df_KorbanJakarta[['wilayah', 'jenis_kelamin']] = df_KorbanJakarta[['wilayah', 'jenis_kelamin']].astype('category')
df_KorbanAnakLakilaki2017['satuan'] = df_KorbanAnakLakilaki2017['satuan'].astype('category')
df_KorbanAnakPerempuan2017['satuan'] = df_KorbanAnakPerempuan2017['satuan'].astype('category') 
df_KorbanJenisKejahatanJawaBarat[['jenis_kekerasan', 'jenis_kelamin_korban', 'satuan']] = df_KorbanJenisKejahatanJawaBarat[['jenis_kekerasan', 'jenis_kelamin_korban', 'satuan']].astype('category')
df_KekerasanTempatKejadianJawaBarat['satuan'] = df_KekerasanTempatKejadianJawaBarat['satuan'].astype('category')
```
